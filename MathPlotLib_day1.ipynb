```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
```

```python
y=[2,3,4,5,6]
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
```

```python
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
```

```python
categories=['A','B','C','D']
values=[10,20,15,30]
plt.bar(categories,values)
# plt.xlabel('X-axis')
# plt.ylabel('Y-axis')
plt.title('Bar Plot')
```

```python
import numpy as np
data=np.random.randn(1000)
```

```python
plt.figure(figsize=(10,20), dpi=200)
```

```python
plt.hist(data,bins=30)
plt.xlabel('values')
plt.ylabel('categories')
```

```python
x=[1,2,3,4,5]
y=[2,3,4,5,6]
a=[2,2,5,4,5]
b=[2,3,7,9,10]
plt.plot(x,y, label='Series 1')
plt.plot(a,b, label='Series 2')
plt.legend(loc='upper right')
```

colors, styles and markers


```python
x=[1,2,3,4,5]
y=[2,3,4,5,6]
a=[2,2,5,4,5]
b=[2,3,7,9,10]
plt.plot(x,y,color='red', label='Series 1',marker='o')
plt.plot(a,b,linestyle='--', label='Series 2')
plt.legend(loc='upper right')
```
